```python
import pandas as pd
```

```python
final_transaction = pd.read_csv('dataset\kf_final_transaction.csv')
inventory = pd.read_csv('dataset\kf_inventory.csv')
cabang = pd.read_csv('dataset\kf_kantor_cabang.csv')
produk = pd.read_csv('dataset\kf_product.csv')
```

```python
tabel_analisa = final_transaction.merge(cabang, how='left', on='branch_id')
tabel_analisa.head()
```

```python
tabel_analisa = tabel_analisa.merge(produk, how='left', on='product_id')
tabel_analisa.head()
```

```python
tabel_analisa.info()
```

```python
def discount(x):
    if x > 500000:
        return 0.3
    elif x > 300000:
        return 0.25
    elif x > 100000:
        return 0.20
    elif x > 50000:
        return 0.15
    else:
        return 0.1

tabel_analisa['presentase_gross_laba'] = tabel_analisa['price_x'].apply(discount)
```

```python
tabel_analisa['nett_sales'] = tabel_analisa['price_x'] - (tabel_analisa['price_x'] * tabel_analisa['discount_percentage'])
tabel_analisa.head()
```

```python
tabel_analisa['nett_profit'] = tabel_analisa['price_x'] * (tabel_analisa['presentase_gross_laba'] - tabel_analisa['discount_percentage'])
tabel_analisa.head()
```

```python
tabel_analisa.info()
```

```python
tabel = tabel_analisa[['transaction_id', 'date', 'branch_id', 'branch_name', 'kota', 'provinsi', 'rating_y', 'customer_name', 'product_id','product_name', 'price_x', 'discount_percentage', 'presentase_gross_laba', 'nett_sales', 'nett_profit', 'rating_x']]
tabel.head()
```

```python
tabel = tabel.rename(columns={
    'rating_y':'rating_cabang',
    'price_x':'actual_price',
    'rating_x':'rating_transaksi'
})
tabel.head()
```

```python
tabel.to_csv('tabel_analisa.csv', index=False)
```
